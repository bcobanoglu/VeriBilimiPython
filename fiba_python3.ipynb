# istatistiksel analizlerdeki temel kavramlar


````python
from statistics import mean, variance, stdev, mode, median, quantiles
from math import sqrt
liste = [0, 1, 1, 2, 2, 3, 3, 3, 3, 4, 5, 6, 7, 8, 2]
print("Aritmetik Ortalama: ", mean(liste))
print("Ortanca deer: ", median(liste))
print("En sk grlen deer(mod): ", mode(liste))
print("Varyans (ortalamadan uzakl): ", variance(liste))
print("Standart Sapma: ", stdev(liste))
print("Standart Sapma edeeri: ", sqrt(variance(liste)))
print("eyrekler..:[Q1:Q2:Q3]", quantiles(liste, n=4))
````

````python
# histogram grafiii
import matplotlib.pyplot as plt

plt.hist(liste, bins=5, alpha=0.6, label='Data Seti')

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Sklk Grafii")
plt.legend()
plt.grid(True)
plt.show()
````

Kovaryans, deikenler (iki veri seti) arasndaki varyans (deiimi) verir. ki deikenin birbiriyle ne kadar deitiini lmek iin kovaryans kullanlr.

Korelasyon, deikenler (iki veri seti) arasndaki ilikiyi ortaya karr.

Korelasyon, iki veri setinin (deikenin) birbiri ile ne kadar ilikili olduunun derecesini ler. Bu ilikinin derecesi ise korelasyon katsays (r) olarak adlandrlr. Korelasyon katsays -1 ile +1 arasnda deien tek bir saydr.

Bu say;

0: ki veri arasnda bir korelasyon (anlaml bir iliki) yoktur.

+1: ki veri arasnda pozitif (ayn ynde) tam (kusursuz) bir korelasyon vardr.

-1: ki veri arasnda negatif (ters ynde) tam (kusursuz) bir korelasyon vardr.

````python
# Bir mterisinin gelir ve giderleri arasndaki ilikiyi grmek iin
from statistics import covariance, correlation
gelir = [5000, 6000, 7000, 8000, 6300, 5800, 6900, 9900, 6500, 4000]
gider = [1000, 3000, 5000, 6000, 8000, 4900, 3000, 7800, 3300, 2000]
print("Kovaryans..:", covariance(gelir, gider))
print("Korelasyon..:", correlation(gelir, gider))
````

````python
# dalm grafii (scatter plot)
import matplotlib.pyplot as plt
plt.scatter(gelir, gider)
plt.xlabel("Gelir")
plt.ylabel("Gider")
plt.title("Mteri Geliri ve Giderleri")
plt.show()
````

````python
# line plot : zamana gre deiim
import matplotlib.pyplot as plt
plt.plot(gelir, label="Gelir")
plt.plot(gider, label="Harcama" )
plt.xlabel("Gelir")
plt.ylabel("Gider")
plt.title("Mteri Geliri ve Giderleri")
plt.show()
````

````python
# Regresyon Grafii: Korelasyonu daha iyi anlamak iin sns.regplot()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setinizi ykleyin.
# rnek veri seti:
data = {'gelir': [5000, 6000, 7000, 8000, 6300, 5800, 6900, 9900, 6500, 4000],
        'gider': [1000, 3000, 5000, 6000, 8000, 4900, 3000, 7800, 3300, 2000]}
df = pd.DataFrame(data)

# Regresyon grafiini izdirin.
sns.regplot(x='gelir', y='gider', data=df)
plt.show()
````

````python
# Aykr deer var m? Dalm simetrik mi? : kutu grafii (boxplot)
plt.boxplot([gelir, gider])
plt.show()
````

````python
# Bir web sayfasndaki bal (title) almak iin
url = "https://docs.python.org/3/"
import requests
from bs4 import BeautifulSoup
r = requests.get(url)  # ierii aldk
if r.status_code == 200:
   soup = BeautifulSoup(r.text, "lxml")
   print(soup.title.string)
else:
    print("Sayfa alnamad")
````
