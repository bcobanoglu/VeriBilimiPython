# istatistiksel analizlerdeki temel kavramlar


````python
from statistics import mean, variance, stdev, mode, median, quantiles
from math import sqrt
liste = [0, 1, 1, 2, 2, 3, 3, 3, 3, 4, 5, 6, 7, 8, 2]
print("Aritmetik Ortalama: ", mean(liste))
print("Ortanca değer: ", median(liste))
print("En sık görülen değer(mod): ", mode(liste))
print("Varyans (ortalamadan uzaklığı): ", variance(liste))
print("Standart Sapma: ", stdev(liste))
print("Standart Sapma eşdeğeri: ", sqrt(variance(liste)))
print("Çeyrekler..:[Q1:Q2:Q3]", quantiles(liste, n=4))
````

````python
# histogram grafiiği
import matplotlib.pyplot as plt

plt.hist(liste, bins=5, alpha=0.6, label='Data Seti')

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Sıklık Grafiği")
plt.legend()
plt.grid(True)
plt.show()
````

Kovaryans, değişkenler (iki veri seti) arasındaki varyansı (değişimi) verir. İki değişkenin birbiriyle ne kadar değiştiğini ölçmek için kovaryans kullanılır.

Korelasyon, değişkenler (iki veri seti) arasındaki ilişkiyi ortaya çıkarır.

Korelasyon, iki veri setinin (değişkenin) birbiri ile ne kadar ilişkili olduğunun derecesini ölçer. Bu ilişkinin derecesi ise korelasyon katsayısı (r) olarak adlandırılır. Korelasyon katsayısı -1 ile +1 arasında değişen tek bir sayıdır.

Bu sayı;

0: İki veri arasında bir korelasyon (anlamlı bir ilişki) yoktur.

+1: İki veri arasında pozitif (aynı yönde) tam (kusursuz) bir korelasyon vardır.

-1: İki veri arasında negatif (ters yönde) tam (kusursuz) bir korelasyon vardır.

````python
# Bir müşterisinin gelir ve giderleri arasındaki ilişkiyi görmek için
from statistics import covariance, correlation
gelir = [5000, 6000, 7000, 8000, 6300, 5800, 6900, 9900, 6500, 4000]
gider = [1000, 3000, 5000, 6000, 8000, 4900, 3000, 7800, 3300, 2000]
print("Kovaryans..:", covariance(gelir, gider))
print("Korelasyon..:", correlation(gelir, gider))
````

````python
# dağılım grafiği (scatter plot)
import matplotlib.pyplot as plt
plt.scatter(gelir, gider)
plt.xlabel("Gelir")
plt.ylabel("Gider")
plt.title("Müşteri Geliri ve Giderleri")
plt.show()
````

````python
# line plot : zamana göre değişim
import matplotlib.pyplot as plt
plt.plot(gelir, label="Gelir")
plt.plot(gider, label="Harcama" )
plt.xlabel("Gelir")
plt.ylabel("Gider")
plt.title("Müşteri Geliri ve Giderleri")
plt.show()
````

````python
# Regresyon Grafiği: Korelasyonu daha iyi anlamak için sns.regplot()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setinizi yükleyin.
# Örnek veri seti:
data = {'gelir': [5000, 6000, 7000, 8000, 6300, 5800, 6900, 9900, 6500, 4000],
        'gider': [1000, 3000, 5000, 6000, 8000, 4900, 3000, 7800, 3300, 2000]}
df = pd.DataFrame(data)

# Regresyon grafiğini çizdirin.
sns.regplot(x='gelir', y='gider', data=df)
plt.show()
````

````python
# Aykırı değer var mı? Dağılım simetrik mi? : kutu grafiği (boxplot)
plt.boxplot([gelir, gider])
plt.show()
````

````python
# Bir web sayfasındaki başlığı (title) almak için
url = "https://docs.python.org/3/"
import requests
from bs4 import BeautifulSoup
r = requests.get(url)  # içeriği aldık
if r.status_code == 200:
   soup = BeautifulSoup(r.text, "lxml")
   print(soup.title.string)
else:
    print("Sayfa alınamadı")
````
